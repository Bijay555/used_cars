```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/cars3.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
# ubstract the year column by year 2023 and rewrite the year column in df

df['year'] = 2023 - df['year']
df.head(25)
```

```python
# Replace 'cylinders ' from string values in the 'cylinders' column
df['cylinders'] = df['cylinders'].apply(lambda x: x.replace(' cylinders', '') if isinstance(x, str) else x)
```

```python
df.head(25)
```

```python
# SInce size has more than half null values we remove it from the dataframe.
df = df.drop('size', axis=1)
```

```python
# It does not make sense to fill the nan values present in year columns. As it is one of major price indicator for used cars. Filling it with synthetic values might mess up the model.
df = df.dropna(subset=['year'])
```

```python
df.shape
```

```python
# understanding dataset stats
df.describe()
```

```python
# list all categorical features

categorical_features = df.select_dtypes(include=['object']).columns.tolist()
print(categorical_features)
```

```python
# splitting the df into train and test

from sklearn.model_selection import train_test_split

train_data, test_data= train_test_split(df, test_size=0.2, random_state=42)

```

```python
train_data.shape
```

```python
test_data.shape
```

```python
from sklearn.impute import SimpleImputer

# Create an instance of SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')  # or any other strategy you prefer

# Fit and transform the training data
train_imputed = imputer.fit_transform(train_data)

# Transform the test data using the same imputer
test_imputed = imputer.transform(test_data)
```

```python
train_imputed[:20]
```

```python
test_imputed[:20]
```

```python
# convert the train_imputed and test_imputed array to dataframe

train_imputed_df = pd.DataFrame(train_imputed, columns=df.columns)
test_imputed_df = pd.DataFrame(test_imputed, columns=df.columns)
```
