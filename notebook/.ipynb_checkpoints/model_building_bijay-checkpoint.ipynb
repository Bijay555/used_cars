```python
# prompt: write a function that takes column and round all the float values to its nearest integer value

import pandas as pd
import numpy as np
def round_to_nearest_int(df, column):
  """
  This function takes a dataframe and a column name and rounds all the float values in that column to its nearest integer value.

  Args:
      df (pd.DataFrame): The dataframe containing the data.
      column (str): The name of the column containing the float values.

  Returns:
      pd.DataFrame: The dataframe with the rounded values in the specified column.
  """

  df[column] = np.round(df[column]).astype(int)
  return df

```

```python
cols_list =['condition', 'cylinders', 'fuel', 'odometer',
       'title_status', 'transmission', 'drive', 'size', 'type', 'paint_color']

for col in cols_list:
  df = round_to_nearest_int(df, column=col)
```

```python
# prompt: split the dataset into train test

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# prompt: build a decision tree regressor and find prediction

from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(X_train, y_train)

predictions = model.predict(X_test)
```

```python
# prompt: calculate mse and r2 score for both traina and test

from sklearn.metrics import mean_squared_error, r2_score

# Calculate MSE for train and test sets
mse_train = mean_squared_error(y_train, model.predict(X_train))
mse_test = mean_squared_error(y_test, predictions)

# Calculate R2 score for train and test sets
r2_train = r2_score(y_train, model.predict(X_train))
r2_test = r2_score(y_test, predictions)

# Print the results
print("MSE Train:", mse_train)
print("MSE Test:", mse_test)
print("R2 Train:", r2_train)
print("R2 Test:", r2_test)
```

#### SHAP (SHapley Additive exPlanations)

SHAP values for a single instance allows us to understand why a specific prediction was made by the model for that instance. SHAP values provide an explanation of the contribution of each feature to the prediction for that particular instance. This explanation helps us understand which features had the most significant impact on the prediction and whether their influence was positive or negative.

```python
import shap
```

```python
# Create a SHAP explainer object
explainer = shap.Explainer(modelL, X_train)
```

```python
# Calculate SHAP values for a single instance (e.g., the first instance in the test set)
shap_values = explainer.shap_values(X_test.iloc[0])
```

```python
# Visualize the SHAP values
shap.initjs()
shap.force_plot(explainer.expected_value, shap_values, X_test.iloc[0], feature_names=X_train.columns)
```

```python
# prompt: perform ann on this and calculate all metrics

import numpy as np
from sklearn.neural_network import MLPRegressor


# Create an MLPRegressor model
model = MLPRegressor(hidden_layer_sizes=(100,), activation='relu', solver='adam', max_iter=1000)

# Train the model
model.fit(X_train, y_train)

# Make predictions on the testing set
predictions = model.predict(X_test)

# Calculate the mean squared error
mse = mean_squared_error(y_test, predictions)

# Calculate the R2 score
r2 = r2_score(y_test, predictions)

# Print the results
print("MSE:", mse)
print("R2:", r2)

# Calculate the mean absolute error
mae = np.mean(np.abs(y_test - predictions))

# Calculate the median absolute error
median_ae = np.median(np.abs(y_test - predictions))

# Print the results
print("MAE:", mae)
print("Median AE:", median_ae)

# Calculate the root mean squared error
rmse = np.sqrt(mse)

# Print the result
print("RMSE:", rmse)
```

```python
# prompt: do kfold split on X and y with k=5 and then build xgboosr model

from sklearn.model_selection import KFold

kf = KFold(n_splits=5)
for train_index, test_index in kf.split(X):
  X_train, X_test = X.iloc[train_index], X.iloc[test_index]
  y_train, y_test = y.iloc[train_index], y.iloc[test_index]

  model_xgb = XGBRegressor()
  model_xgb.fit(X_train, y_train)

  predictions_xgb = model_xgb.predict(X_test)

  # Calculate MSE for train and test sets
  mse_train_xgb = mean_squared_error(y_train, model_xgb.predict(X_train))
  mse_test_xgb = mean_squared_error(y_test, predictions_xgb)

  # Calculate R2 score for train and test sets
  r2_train_xgb = r2_score(y_train, model_xgb.predict(X_train))
  r2_test_xgb = r2_score(y_test, predictions_xgb)

  # Print the results
  print("MSE Train (XGBoost):", mse_train_xgb)
  print("MSE Test (XGBoost):", mse_test_xgb)
  print("R2 Train (XGBoost):", r2_train_xgb)
  print("R2 Test (XGBoost):", r2_test_xgb)
```

```python
# prompt: implement a ensemble learning

from sklearn.ensemble import VotingRegressor

# Define the individual models
model1 = DecisionTreeRegressor()
model2 = RandomForestRegressor()
model3 = XGBRegressor()

# Create the ensemble model
ensemble_model = VotingRegressor(estimators=[('dt', model1), ('rf', model2), ('xgb', model3)])

# Train the ensemble model
ensemble_model.fit(X_train, y_train)

# Make predictions using the ensemble model
predictions_ensemble = ensemble_model.predict(X_test)

# Evaluate the ensemble model
mse_ensemble = mean_squared_error(y_test, predictions_ensemble)
r2_ensemble = r2_score(y_test, predictions_ensemble)

# Print the results
print("MSE Ensemble:", mse_ensemble)
print("R2 Ensemble:", r2_ensemble)

```
